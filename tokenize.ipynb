```python
## source: https://stackoverflow.com/questions/36460472/custom-word-tokenizer

tweet = u""+df.loc[0].text

emoticons = r'(?:[:;=\^\-oO][\-_\.]?[\)\(\]\[\-DPop_\^\\\/])'

regex_tweets = [
    emoticons,
    r'<[^>]+>',      ## HTML TAGS
    r'(?:@[\w\d_]+)',   ## @-mentions
    r'(?:\#[\w]+)',  ## #HashTags
    r'(?:[\.,]{2,})', ## repeated . and ,
    # r'(?:[A-Z]{1,3}[a-z]{0,10}[,\.]+[A-Z]?[a-z]{0,3}[,\.]?){1,3}', ## Titles
    r'Rp|IDR|USD|SGD|JPY|SAR|AUD', ## currency
    r'http[s]?://(?:[a-z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-f][0-9a-f]))+', # URLs
    r"(?:[A-z][A-z'\-_]+[A-z]?)", # words with - and '
    r'(?:(?:\d+,?)+(?:\.?\d+)?)', ##numbers
    r'(?:[\w_]+)',   #other words
    r'(?:\S)'        ## normal text 
]

#compiling regex
tokens_re = re.compile(r'('+'|'.join(regex_tweets)+')' , re.VERBOSE)
# tokens_re.findall(tweet)
```
