```python
df = pd.read_csv('Sample Tweets\sample_tweet-brand2.csv')
sample = df.sample(df.shape[0])
sample
```

```python
import nltk
```

```python
tw=nltk.TweetTokenizer()
sample['tokenized']=sample.text.apply(tw.tokenize)
```

```python
sample[['id','text']].to_csv('sample.txt', header=None, index=None, sep=' ', mode='a')
```

```python
from random import choice

with open('sample.txt', 'r',  encoding="mbcs") as r:
    lines = r.readlines()

with open('sample_shuffled_1.txt', 'w', encoding="mbcs") as w:
    while lines:
        l = choice(lines)
        lines.remove(l)
        w.write(l)
```
